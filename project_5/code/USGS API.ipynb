```python
import pandas as pd
import requests
import time
```

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
```

Running the API to get the last 20 years of earthquake information for everything above magnitude 3.  The longitude and latitude limit it to the area around the lower 48 states.

```python
params = {
    'format' : 'geojson',
    'starttime' : '2000-01-01',
    'endtime' : '2020-07-30',
    'limit' : 20000,
    'minmagnitude' : 3,
    'eventtype' : 'earthquake',
    'minlatitude' : 24.6,
    'minlongitude' : -125,
    'maxlatitude' : 50,
    'maxlongitude' : -65
    }
res = requests.get(url, params)    
    
   
   
    
```

```python
data = res.json()
```

converting the data to a usable state

```python
events = data['features']
```

```python
df1 = pd.DataFrame(events)
```

```python
df1
```

```python
# code help for turning the dictionary in the column to seperate columns 
#https://stackoverflow.com/questions/38231591/splitting-dictionary-list-inside-a-pandas-column-into-separate-columns
df2 = pd.json_normalize(df1['properties'])
```

```python
df2.rename(columns={'type':'earthquake' }, inplace=True)
df2
```

```python
df3 = pd.json_normalize(df1['geometry'])
```

```python
df3
```

```python
# code help https://stackoverflow.com/questions/35491274/pandas-split-column-of-lists-into-multiple-columns
# seperating the list to columns
df3[['long','lat', 'depth']] = pd.DataFrame(df3.coordinates.tolist(), index= df3.index)
df3.rename(columns={'type':'point'}, inplace=True)
df3
```

```python
# fixing the columns in the data frames before the merge
df1.drop(columns = ['properties', 'geometry'], inplace=True)
df3.drop(columns = 'coordinates', inplace=True)
```

```python
df_final = pd.concat([df1,df2,df3], axis=1)
```

```python
df_final
```

```python
df_final.to_csv('./../data/USGS/earthquake.csv', index=False)
```
